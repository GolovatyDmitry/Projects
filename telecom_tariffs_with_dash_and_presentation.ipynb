# Поиск клиентов с неоптимальными тарифами

## Введение

**Задачи:**

- Найти клиентов с неоптимальным тарифом;
- Изучить изменение выручки при изменении тарифов клиентам на оптимальные;
- Предложить решение: как уменьшить риск падения выручки.

Датасеты, представленные для исследования, содержат данные об использовании услуг виртуальной телефонии.

**Правила формирования тарифов:**

Ежемесячная плата за тариф:

- Тариф А — 5000 ₽,
- B — 2000 ₽,
- С — 1000 ₽.

Входящие звонки — бесплатны на всех тарифах,

Исходящие звонки внутри сети — бесплатны до 2000 минут в месяц на всех тарифах. Цена сверх лимита:

- A — 0.1 ₽/мин.,
- B — 0.15 ₽/мин.,
- C — 0.3 ₽/мин.

Исходящие звонки, кроме внутренних:

- A — 0.4 ₽/мин.,
- B — 0.5 ₽/мин.,
- C — 0.7₽/мин.

За каждого оператора клиент платит 100 ₽ в месяц на любом тарифе — дополнительно к стоимости тарифа и цене звонков.

Счета выставляются на прошедший календарный месяц. Продолжительность звонков округляется в большую сторону до минуты. Время ожидания ответа также входит в тарификацию.

Однако необходимо обратить внимание колег на присутствие пропусков в столбце `internal` и разобраться в их истинном происхождение.

### Проверка единицы измерения длительности звонков

Изначально есть предположение, что длительность звонков представлена в секундах. Однако, во-первых, у нас есть очень много значений с длительностью до 1000 секунд, что подозрительно мало, во-вторых, есть значения свыше 43200 сек (что больше 12 часового рабочего дня), что вызывает еще большие сомнения.

Рассмотрим длительность звонков в течение времени.

```python
#Группировка длительности звонков по дням
df_day_duration_mean = df_telecom.groupby('date').agg(
    {'total_call_duration':'mean', 'calls_count':'mean','user_id':'first'}).reset_index()
```

```python
#График длительности звонков по дням
df_day_duration_mean.plot(x='date', y='total_call_duration', figsize=(10,5))
plt.title('Изменение средней длительности звонков по дням')
plt.xlabel('Дата')
plt.ylabel('Длительность звонков')
plt.show()

#График количества звонков по дням
df_day_duration_mean.plot(x='date', y='calls_count', figsize=(10,5))
plt.title('Изменение среднего количества звонков по дням')
plt.xlabel('Дата')
plt.ylabel('Количество звонков')
plt.show()
```

```python
#График средней длительности звонков по клиентам
df_user_duration.query('total_call_duration > 600').plot(x='user_id', y='total_call_duration', kind='bar', figsize=(18,8))
plt.title('Изменение средней длительности звонков по клиентам (длительность более 600 с)')
plt.xlabel('Клиент')
plt.ylabel('Длительность звонков')
plt.show()
```

```python
#Топ 10 по средней длительности звонков среди клиентов
df_user_duration.sort_values('total_call_duration', ascending=False).head(10)
```

### Затраты клиентов на услуги

```python
df_user_duration = df.groupby(['tariff_plan', 'month_activity', 'user_id', 'direction', 'internal']).agg(
{'total_call_duration':'sum'}).reset_index()
# #Убираем входящие звонки, так как они не тарифицируются
# df_user_costs = df_user_costs.query('direction != "in"')
#Создаем столбцец с длительность звонков в минутах
df_user_duration['total_call_duration_min'] = np.ceil(df_user_duration['total_call_duration']/60).astype('int')
df_user_duration.head(10)
```

```python
#Количество операторов в месяц у каждого клиента
df_operators_count = df[~df['operator_id'].isna()].groupby(['tariff_plan', 'month_activity', 'user_id'])['operator_id'].nunique().reset_index()
```

```python
#ПАРАМЕТРЫ ТАРИФА

#Абонентская плата
fee_A = 5000
fee_B = 2000
fee_C = 1000

#Исходящие внутри сети (до 2000 минут бесплатны)
out_internal_A = 0.1
out_internal_B = 0.15
out_internal_C = 0.3

#Исходящие внешние звонки
out_external_A = 0.4
out_external_B = 0.5
out_external_C = 0.7

#Лимит бесплатных минут исходяих внутри сети
out_internal_free_min = 2000

#Плата за одного оператора
operators_fee = 100

df_tariff = pd.DataFrame([
    ['A', fee_A, out_internal_A, out_external_A, out_internal_free_min, operators_fee],
    ['B', fee_B, out_internal_B, out_external_B, out_internal_free_min, operators_fee],
    ['C', fee_C, out_internal_C, out_external_C, out_internal_free_min, operators_fee]],
    columns=['tariff', 'fee', 'out_internal', 'out_external', 'out_internal_free_min', 'operators_fee']).set_index('tariff')
df_tariff
```

```python

def cost_calculation(row):  
    '''
Функция подсчета затрат клиентов, исходя только из длительности звонков их вида и направления. 
Длительность должна быть просуммирована с группировкой по тарифу, месяцу, направлению и виду.
Параметры тарифного плана берутся из таблицы df_tariff в зависимости от тарифа
    '''
    
    if row['tariff_plan'] == 'A':
        if row['direction'] == 'out':
            if row['internal'] == False:
                costs = row['total_call_duration_min'] * df_tariff.loc['A','out_external']
            elif row['total_call_duration_min'] > 2000:
                costs = row['total_call_duration_min'] * df_tariff.loc['A','out_internal']
            else:
                costs = 0
        else:
            costs = 0           
    elif row['tariff_plan'] == 'B':
        if row['direction'] == 'out':
            if row['internal'] == False:
                costs = row['total_call_duration_min'] * df_tariff.loc['B','out_external']
            elif row['total_call_duration_min'] > 2000:
                costs = row['total_call_duration_min'] * df_tariff.loc['B','out_internal']
            else:
                costs = 0
        else:
            costs = 0   
    else:
        if row['direction'] == 'out':
            if row['internal'] == False:
                costs = row['total_call_duration_min'] * df_tariff.loc['C','out_external']
            elif row['total_call_duration_min'] > 2000:
                costs = row['total_call_duration_min'] * df_tariff.loc['C','out_internal']
            else:
                costs = 0 
        else:
            costs = 0   
    return costs           
```

```python
#Применяем функцию к таблице, расчитываем расходы клиентов по звонкам
df_user_duration['call_costs'] = df_user_duration.apply(cost_calculation, axis = 1)
df_user_duration.tail()
```

```python
#Группируем таблицу по тарифу, месяцу, клиенту, чтобы понять общие затраты по звонкам
df_user_costs = df_user_duration.groupby(['tariff_plan', 'month_activity', 'user_id']).agg(
    {'call_costs':'sum'}).reset_index()
df_user_costs.head()
```

```python
#Объединяем таблицы с затратами и количеством операторов в месяц по клиентам
df_user_costs = df_operators_count.merge(df_user_costs, how='right', on=['tariff_plan','month_activity', 'user_id'])
df_user_costs = df_user_costs.rename(columns={'operator_id':'operator_count'})
df_user_costs.head()
```

```python
#Заменяем пропуски, образованные из-за нулевого количества операторов на нули
df_user_costs['operator_count'] = df_user_costs['operator_count'].fillna(0)
```

```python
def total_costs_calsulation(row):
    '''
Функция для расчета общих затрат на тариф для клиента в месяц.
Складываются затраты на операторов, абонентскую плату и звонки.
Исходная таблица должна быть сгруппирована по тарифу, месяцу и клиенту 
    и иметь количество операторов и расходы на звонки.
Абонентская плата из таблицы df_tariff.
    '''    
    operator_costs = row['operator_count'] * 100
    
    if row['tariff_plan'] == 'A':
        fee = df_tariff.loc['A','fee']
        
    elif row['tariff_plan'] == 'B':
        fee = df_tariff.loc['B','fee']
        
    else:
        fee = df_tariff.loc['C','fee']
    
    total_costs = operator_costs + fee + row['call_costs']
    return total_costs  
    
```

```python
#Применяем функцию к таблице, расчитываем общие расходы клиентов
df_user_costs['total_costs'] = df_user_costs.apply(total_costs_calsulation, axis = 1)
df_user_costs.head()
```

```python
'''
Функции подсчета затрат клиентов на каждом тарифе, исходя только из длительности звонков их вида и направления. 
Длительность должна быть просуммирована с группировкой по тарифу, месяцу, направлению и виду.
Параметры тарифного плана берутся из таблицы df_tariff в зависимости от тарифа
'''
#Функция подсчета затрат клиентов на тарифе А
def cost_calculation_A(row):   
    if row['direction'] == 'out':
        if row['internal'] == False:
            costs = row['total_call_duration_min'] * df_tariff.loc['A','out_external']
        elif row['total_call_duration_min'] > 2000:
            costs = row['total_call_duration_min'] * df_tariff.loc['A','out_internal']
        else:
            costs = 0
    else:
        costs = 0           
    return costs 

#Функция подсчета затрат клиентов на тарифе B
def cost_calculation_B(row):     
    if row['direction'] == 'out':
        if row['internal'] == False:
            costs = row['total_call_duration_min'] * df_tariff.loc['B','out_external']
        elif row['total_call_duration_min'] > 2000:
            costs = row['total_call_duration_min'] * df_tariff.loc['B','out_internal']
        else:
            costs = 0
    else:
        costs = 0
    return costs     

#Функция подсчета затрат клиентов на тарифе C
def cost_calculation_C(row):    
    if row['direction'] == 'out':
        if row['internal'] == False:
            costs = row['total_call_duration_min'] * df_tariff.loc['C','out_external']
        elif row['total_call_duration_min'] > 2000:
            costs = row['total_call_duration_min'] * df_tariff.loc['C','out_internal']
        else:
            costs = 0 
    else:
        costs = 0   
    return costs  
```

```python
df_duration_all_tariffs = df_user_duration.drop(columns='call_costs').copy()
df_duration_all_tariffs['costs_a'] = df_duration_all_tariffs.apply(cost_calculation_A, axis = 1)
df_duration_all_tariffs['costs_b'] = df_duration_all_tariffs.apply(cost_calculation_B, axis = 1)
df_duration_all_tariffs['costs_c'] = df_duration_all_tariffs.apply(cost_calculation_C, axis = 1)
df_duration_all_tariffs.tail()
```

```python
#Группируем таблицу по тарифу, месяцу, клиенту, чтобы понять общие затраты по звонкам
df_user_costs_all_tariffs = df_duration_all_tariffs.groupby(['tariff_plan', 'month_activity', 'user_id']).agg(
    {'costs_a':'sum', 'costs_b':'sum', 'costs_c':'sum'}).reset_index()


#Объединяем таблицы с затратами и количеством операторов в месяц по клиентам
df_user_costs_all_tariffs = df_operators_count.merge(df_user_costs_all_tariffs, how='right', on=['tariff_plan','month_activity', 'user_id'])
df_user_costs_all_tariffs = df_user_costs_all_tariffs.rename(columns={'operator_id':'operator_count'})
df_user_costs_all_tariffs.head()
```

```python
#Заменяем пропуски, образованные из-за нулевого количества операторов на нули
df_user_costs_all_tariffs['operator_count'] = df_user_costs_all_tariffs['operator_count'].fillna(0)
```

```python
'''
Функции для расчета общих затрат клиента в месяц для каждого тарифа.
Складываются затраты на операторов, абонентскую плату и звонки.
Исходная таблица должна быть сгруппирована по тарифу, месяцу и клиенту 
    и иметь количество операторов и расходы на звонки.
Абонентская плата из таблицы df_tariff.
'''
#Функция для тарифа А
def total_costs_calsulation_A(row):
    operator_costs = row['operator_count'] * 100
    fee = df_tariff.loc['A','fee']
    total_costs = operator_costs + fee + row['costs_a']
    return total_costs 

#Функция для тарифа B
def total_costs_calsulation_B(row):
    operator_costs = row['operator_count'] * 100
    fee = df_tariff.loc['B','fee']
    total_costs = operator_costs + fee + row['costs_b']
    return total_costs

#Функция для тарифа C
def total_costs_calsulation_C(row):
    operator_costs = row['operator_count'] * 100
    fee = df_tariff.loc['C','fee']
    total_costs = operator_costs + fee + row['costs_c']
    return total_costs  
    
```

```python
df_user_costs_all_tariffs['total_costs_a'] = df_user_costs_all_tariffs.apply(total_costs_calsulation_A, axis = 1)
df_user_costs_all_tariffs['total_costs_b'] = df_user_costs_all_tariffs.apply(total_costs_calsulation_B, axis = 1)
df_user_costs_all_tariffs['total_costs_c'] = df_user_costs_all_tariffs.apply(total_costs_calsulation_C, axis = 1)
df_user_costs_all_tariffs.head()
```

```python
#Функция для выбора оптимального тарифа
def select_optimal_tarif(row):
    if row['total_costs_a'] == min(row['total_costs_a'], row['total_costs_b'], row['total_costs_c']):
        optimal = 'A'
    elif row['total_costs_b'] == min(row['total_costs_a'], row['total_costs_b'], row['total_costs_c']):
        optimal = 'B'
    else:
        optimal = 'C'
    return optimal    
```

```python
#Функция для выбора текущих трат по текущиму тарифу из таблицы
def current_cost(row):
    if row['tariff_plan'] == 'A':
        current_costs = row['total_costs_a']
    elif row['tariff_plan'] == 'B':
        current_costs = row['total_costs_b']
    else:
        current_costs = row['total_costs_c']
    return current_costs

#Функция для выбора оптимальных трат из таблицы
def optimal_cost(row):
    optimal_costs = min(row['total_costs_a'], row['total_costs_b'], row['total_costs_c'])
    return optimal_costs
```

## Проверка статистических гипотез

Сформулируем и проверим несколько гипотез.

### Средняя выручка с клиентов до и после перевода на оптимальные тарифы различается

Наша задача - узнать различаются ли средние месячные выручки с клиентов до и после перевода на оптимальные тарифы.

Сформулируем нулевую гипотезу(H0): средние месячные выручки до и после перевода на оптимальные тарифы равны между собой.

Альтернативная гипотеза (H1): средние месячные выручки до и после перевода на оптимальные тарифы различаются.

Оценивать случайность полученых нами результатов будем с помощью p-значения. Уровень значимости зададим как 5%. Если p-значение будет меньше 5% - у нас будут основания отвергнуть нулевую гипотезу. Если p-значение будет больше 5%, то оснований отвергнуть нулевую гипотезу нет.

Проверим тип распределения наших значений:

```python
#Средние выручки с клиентов в месяц 
df_month_user_mean = df_user_costs.groupby('user_id')['total_costs'].mean().reset_index()

#Средние выручки с клиентов в месяц после перевода на оптимальный тариф
df_month_user_mean_optimal = df_user_costs_all_tariffs.groupby('user_id')['optimal_costs'].mean().reset_index()
```

```python
#Критический уровень статистической значимости
alpha = 0.05
```

```python
#Проверяем нормальность распределения
results = st.shapiro(df_month_user_mean['total_costs'])
p_value = results[1]

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
```

```python
#Проверяем нормальность распределения
results = st.shapiro(df_month_user_mean_optimal['optimal_costs'])
p_value = results[1]

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
```

Статистическую значимость различий конверсий будем считать, используя тест Манна-Уитни, так как распределение выборок далеко от нормального.

```python
#Сравнение средних
print("P-value: {0:.5f}".format(st.mannwhitneyu(
    df_month_user_mean['total_costs'], df_month_user_mean_optimal['optimal_costs'])[1]))
```

**P-value меньше нашего уровня значимости, что позволяет отвергнуть нулевую гипотезу о равенстве средних выручек до и после перевода на оптимальные тарифы.**
