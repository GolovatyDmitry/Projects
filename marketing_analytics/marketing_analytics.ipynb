Задача: помочь маркетологам Яндекс.Афиши снизить расходы — отказаться от невыгодных источников трафика и перераспределить бюджет.

## Продуктовые метрики

### Рассчет DAU, WAU и MAU.

```python
#Создаем столбцы с годом, месяцем, неделей и датой старта сессии
df_visits['year'] = df_visits['start_time'].dt.year
df_visits['week'] = df_visits['start_time'].dt.week
df_visits['date'] = df_visits['start_time'].dt.date
df_visits['first_date_month'] =df_visits['start_time'].astype('datetime64[M]')
```

```python
#Создаем функцию для построения графиков
def plotgraph(data, y, x, title, ylabel, xlabel, figsize=(13,4)):
    data.plot(figsize=figsize, y=y, x=x, title=title, grid=True)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.show()
```

```python
#Создаем сгрупированную таблицу по первому дню месяца и расчитываем MAU
df_mau = df_visits.groupby(['first_date_month'])['uid'].nunique().reset_index()
mau = int(df_mau['uid'].mean())
display('MAU за весь период = {}'.format(mau))

#Строим график изменения mau во времени
plotgraph(df_mau, 'uid', 'first_date_month', 'Изменение MAU во времени', 'Количество уникальных пользователей в месяц (MAU)', 'Дата')
```

```python
#Создаем сгрупированную таблицу по году и неделе и расчитываем WAU
df_wau = df_visits.groupby(['year', 'week']).agg({'uid':'nunique', 'date':'min'}).reset_index()
wau = int(df_wau['uid'].mean())
display('WAU за весь период = {}'.format(wau))

#Строим график изменения WAU во времени
plotgraph(df_wau, 'uid', 'date', 'Изменение WAU во времени', 'Количество уникальных пользователей в неделю (WAU)', 'Дата (год-месяц)')
```

```python
#Создаем сгрупированную таблицу по дате и расчитываем DAU
df_dau = df_visits.groupby(['date'])['uid'].nunique().reset_index()
dau = int(df_dau['uid'].mean())
display('DAU за весь период = {}'.format(dau))

#Строим график изменения DAU во времени
plotgraph(df_dau, 'uid', 'date', 'Изменение DAU во времени', 'Количество уникальных пользователей в день (DAU)', 'Дата (год-месяц)')
```

**На графиках прослеживается повышенная активность пользователей с августа по декабрь и спад спроса с декабря по май. Это может быть связано с сезонностью бизнеса, что требует проверки на большем временном периоде. Также это может быть обусловлено рекламной политикой.**

### Расчет Retention Rate

Сгруппируем когорты по месяцам первого захода пользователей на сайт

```python
#Создаем столбец с датой первого захода пользователя 
df_visits['first_activity_date'] = df_visits.groupby(['uid'])['date'].transform('min')
```

```python
#Создаем столбцы с месяцами (первый день месяца) первого захода и каждой активности пользователя
df_visits['first_activity_month'] = df_visits['first_activity_date'].astype('datetime64[M]')
df_visits['activity_month'] = df_visits['date'].astype('datetime64[M]')

#Создаем столбец с lifetime когорты
df_visits['cohort_lifetime'] = (
    ((df_visits['activity_month'] - df_visits['first_activity_month']) / 
    np.timedelta64(1,'M')).round().astype('int') 
)

#Создаем таблицу с количеством уникальных пользователей в lifetime каждой когорты
cohorts = df_visits.groupby(['first_activity_month', 'cohort_lifetime'])['uid'].nunique().reset_index()

#Создаем таблицу с количеством пользователей в нулевой lifetime когорте
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_month', 'uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})

#Объединяем таблицы
cohorts = cohorts.merge(initial_users_count, on='first_activity_month')

#Cчитаем в новом столбце показатель Retention Rate
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users']

retention_pivot = cohorts.pivot_table(
    index='first_activity_month', columns='cohort_lifetime', values='retention', aggfunc='sum')
retention_pivot
```

Расчитаем средний Retention Rate на второй месяц жизни когорт.

```python
retention_2month = retention_pivot[1].mean()
display('Средний Retention Rate на второй месяц жизни когорт: {}'.format(round(retention_2month,3)))
```

Построим тепловую карту для наглядного отображения измнения Retention Rate по времени "жизни" когорты.

```python
#удалим столбец с первым месяцем жизни когорты для наглядности
retention_pivot_map = retention_pivot.drop(columns=0)

cohorts_date = retention_pivot_map.index.to_series().dt.strftime('%Y-%m-%d')
plt.figure(figsize=(13, 9))
plt.title('Изменение Retention Rate по времени "жизни" когорты')
sns.heatmap(retention_pivot_map, annot=True, fmt='.2%', linewidths=1, linecolor='black', vmin=0.00, yticklabels = cohorts_date)
plt.show()
```

### Время до первой покупки

```python
#создаем Series с временем первой покупки пользователем и добавляем эту информацию в df_orders
time_first_order = df_orders.groupby(['uid'])['buy_time'].min()
time_first_order.name = 'time_first_order'
df_orders = df_orders.join(time_first_order, on='uid')
```

```python
#создаем Series с временем первого визита пользователя
time_first_visit = df_visits.groupby(['uid'])['start_time'].min()
time_first_visit.name = 'time_first_visit'
```

```python
#объединяем df_first_actions и time_first_order
df_first_actions = time_first_order.reset_index().join(time_first_visit, on='uid', how='inner')


#создаем столбец с разницей в часах между датой первой покупки и первого визита
df_first_actions['time_before_buying_hour'] = (
    (df_first_actions['time_first_order'] - df_first_actions['time_first_visit']) / np.timedelta64(1,'h'))
```

```python
df_first_actions.head()
```

```python
df_first_actions['time_before_buying_hour'].hist()
plt.title('Распределение времени до первой покупки')
plt.ylabel('Частота')
plt.xlabel('Продолжительность, ч')
plt.show()
```

```python
display('Наиболее частое (мода) время с первого визита до покупки: {} часов'.format(df_first_actions['time_before_buying_hour'].mode().round(2)[0]))
```

```python
display('Среднее время (медиана) с первого визита до покупки: {} часов'.format(round(df_first_actions['time_before_buying_hour'].median(), 2)))
```

### Среднее количество покупок

Расчитаем среднее количество покупок на одного покупателя за последние 6 месяцев (декабрь 2017г. - май 2018г.).

```python
#Выделяем в отдельном столбце первый день месяца покупки
df_orders['month_orders'] = df_orders['buy_time'].astype('datetime64[M]')

#Считаем кол-во покупок и кол-во уникальных пользователей, сгрупированных по месяцам
count_orders = df_orders.groupby(['month_orders']).agg({'revenue':'count', 'uid':'nunique'}).reset_index()
count_orders = count_orders.rename(columns={'revenue':'quantity_orders', 'uid':'unique_uid'})
#Считаем в новом столбце кол-во покупок на одного покупателя
count_orders['orders_per_uid'] = count_orders['quantity_orders'] / count_orders['unique_uid']

display('Среднее количество покупок на одного покупателя за последние 6 месяцев: {}'.format(round(count_orders.loc[6:11]['orders_per_uid'].mean(),2)))
```

### Средний чек

```python
#Группируем заказы по времени и считаем сумму и кол-во заказов
check = df_orders.groupby(['buy_time']).agg({'revenue':['sum', 'count']}).reset_index()
check.columns = ['buy_time', 'sum_revenue', 'quantity_purchase']
#В новом столбце расчитываем средний чек за конкретное время и выводим общее среднее
check['check_mean'] = check['sum_revenue'] / check['quantity_purchase']
display ('Средний чек за исследуемый период: {} у.е.'.format(round(check['check_mean'].mean(),2)))
```

```python
#Для отображения динамики среднего чека во времени найдем средние за месяц и отобразим их на графике
check['first_day_month'] = check['buy_time'].astype('datetime64[M]')
check_by_month = check.groupby('first_day_month')['check_mean'].mean().reset_index()
plotgraph(check_by_month, 'check_mean', 'first_day_month', 'Изменение среднего чека во времени', 'Сумма чека', 'Дата')
```

**Из графика можем видеть, что максимальный размер чека устанавливается в декабре и марте. Возможно это связано с тем, что люди в это время покупают продукты сайта в подарок. 
Однако, также как и на графике с динамикой посещения сайта, мы видим небольшое снижение размера чека с декабря.**

### LTV на покупателя по когортам

Рассчитаем средний LTV по когортам за 6 месяцев. В расчёт включим когорты, «прожившие» не менее 6 месяцев. Маржинальность сервиса принимаем за 100%.

```python
#Добавляем в таблицу месяц первой покупки 
df_orders['first_order_month'] = df_orders['time_first_order'].astype('datetime64[M]')

#Создаем таблицу с размером (кол-вом уникальных пользователей) когорт
cohort_sizes = (
    df_orders.groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_order_month', 'n_buyers']

#Маржинальность по условия 100%
margin_rate = 1

#Создаем таблицу с суммарными доходами по когортам  и месяцам покупок
cohorts = (
    df_orders.groupby(['first_order_month', 'month_orders'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
#Объединяем таблицы cohort_sizes и cohorts
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
#Добавляем столбец с прибылью на когорту
report['gp'] = report['revenue'] * margin_rate
#Добавляем столбец с возрастом когорты в месяцах на момент покупки 
report['age'] = (
    report['month_orders'] - report['first_order_month']
) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
#Добавляем столбец с LTV когорты
report['ltv'] = report['gp'] / report['n_buyers']
#Результирающая таблица с LTV
result = report.pivot_table(
    index='first_order_month',
    columns='age',
    values='ltv',
    aggfunc='mean'
)
```

```python
#Строим тепловую карту изменения куммулятивной LTV
cohorts_date = result.index.to_series().dt.strftime('%Y-%m-%d')
plt.figure(figsize=(15, 10))
sns.heatmap(result.cumsum(axis=1), annot=True, fmt='.2f', linewidths=1, linecolor='black', yticklabels=cohorts_date)
plt.title('Изменение LTV в когортах с "возрастом" когорты')
plt.xlabel('Возраст когорты')
plt.ylabel('Когорты')
plt.show()
```

```python
#Рассчет среднего LTV по когортам за 6 месяцев с возрастом >= 6.
m6_cum_ltv = result.cumsum(axis=1).mean(axis=0)
display('Cредний LTV за 6 месяцев когорт с возрастом более 5 месяцев: {} у.е.'.format(round(m6_cum_ltv[5:11].mean(),2)))
```

### Выводы

- Среднее время с первого визита до первой покупки: 405.67 часов;
- среднее количество покупок на одного покупателя за последние 6 месяцев: 1.23, что в подтверждение низкого Retetnion Rate говорит о том, что клиенты делают в основном одну покупку и редко возвращаются за второй;
- средний чек за исследуемый период: 4.99 у.е.
- при изучении LTV можно выделить когорты пользователей, начавших пользоваться сайтом в июне и сентябре 2017 года; они приносят наибольшую часть выручки;
- средний LTV за 6 месяцев когорт с возрастом более 5 месяцев: 9.13 у.е.

```python
#Добавляем столбец с временем первого визита в таблицу с визитами
df_visits = df_visits.join(time_first_visit, on='uid')

#Находим первые источники визитов для каждого пользователя
main_source_id = df_visits.query('start_time == time_first_visit').groupby('uid')['source_id'].first().reset_index()
main_source_id.head()
```

### ROMI по когортам в разрезе источников

```python
#Считаем сумму дохода по каждой когорте и источнику
df_result = df_orders.groupby(['first_order_month','month_orders', 'source_id'])['revenue'].sum().reset_index()
```

```python
#Считаем возраст когорты и валовую прибыль
df_result['age'] = ((df_result['month_orders'] - df_result['first_order_month']) / np.timedelta64(1, 'M')).round().astype('int')
df_result['gp'] = df_result['revenue'] * margin_rate
```

```python
#Объединяем таблицу с таблицей месячных расходов и считаем LTV
df_result = df_result.merge(monthly_costs, on=['first_order_month', 'source_id'])
df_result['ltv'] = df_result['gp'] / df_result['n_buyers']
```

```python
#Строим сводную таблицу с отображением romi по когортам и возрасту 
romi_by_age = df_result.pivot_table(index='first_order_month', columns='age', values='romi', aggfunc='mean').cumsum(axis=1).round(2)


plt.figure(figsize=(10, 7))
sns.heatmap(romi_by_age, annot=True, fmt='.2f', linewidths=1, linecolor='black', yticklabels=cohorts_date)
plt.title('Изменения ROMI по когортам с "возрастом" когорты')
plt.xlabel('Возраст')
plt.ylabel('Когорты')
plt.show()
```
